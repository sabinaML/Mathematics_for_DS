# Посчитайте частные производные функций:

```python
import sympy as sym
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
x=sym.Symbol('x')
y=sym.Symbol('y')
```

```python
f=2*x**2*y**3+1/x+y**2*x+7 #1
```

```python
sym.diff(f,x) # частная производная по x
```

```python
sym.diff(f,y) # частная производная по y
```

```python
f = x**2*y-sym.sin(x*y) +sym.cos(x**2) +6*y #2
```

```python
sym.diff(f,x) # частная производная по x
```

```python
sym.diff(f,y) # частная производная по y
```

# Вычислите градиент функции в точке (1,2)

```python
f=2*x**2*y**3+1/x+y**2*x+7 
```

```python
np.array([sym.diff(f,x).subs(x,1).subs(y,2),sym.diff(f,y).subs(x,1).subs(y,2)]) # градиент функции в точке (1,2)
```

# Постройте трехмерный график функции

```python
 from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter

fig=plt.figure()
ax=fig.gca(projection='3d')

X=np.arange(-50,50,0.25)
Y=np.arange(-50,50,0.25)
X,Y = np.meshgrid(X,Y)
Z=Y**2-X**2*Y+2*X*Y
surf=ax.plot_surface(X,Y,Z, rcount=20, ccount=20, cmap='jet',
                      linewidth=0.5, edgecolors='k')
plt.show()
```
